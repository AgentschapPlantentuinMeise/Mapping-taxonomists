```python
import pandas as pd
import numpy as np
import pickle
```

```python
taxon = pd.read_csv("./backbone/Taxon.tsv", sep="\t", on_bad_lines='skip')
taxon
```

```python
animals = taxon[taxon["kingdom"]=="Animalia"]
plants = taxon[taxon["kingdom"]=="Plantae"]
fungi = taxon[taxon["kingdom"]=="Fungi"]
```

```python
insects = animals[animals["class"]=="Insecta"].reset_index(drop=True)
insects
```

```python
accepted_insects = insects[insects["taxonomicStatus"]=="accepted"].reset_index(drop=True)
insect_species = accepted_insects[np.logical_not(accepted_insects["specificEpithet"].isnull())].reset_index()
insect_species
```

```python
insect_species.columns
```

```python
# filter sub-specific variations
insect_species_only = insect_species[["order", "family", "genus", "specificEpithet"]]
insect_species_only = insect_species_only.drop_duplicates().reset_index(drop=True)
insect_species_only
```

```python
species_per_order = {}

for order in set(insect_species_only["order"]):
    species_per_order[order] = len(insect_species_only[insect_species_only["order"]==order].index)

species_per_order
```

```python
insect_articles = pd.read_pickle("./data/insect_articles_search.pkl")
insect_articles
```

```python
# convert to RLS
order_rls = {}

for order in set(insect_articles["order"]):
    art_count = len(insect_articles[insect_articles["order"]==order].index)
    
    if order != "Isoptera": # actually a subclass...
        order_rls[order] =  art_count/species_per_order[order]*100

order_rls
```

```python
# filter for European articles based on first institution of first author
eu_codes = ["AT", "BE", "BG", "HR", "CY", "CZ", "DK", "EE", "FI", "FR", "DE", "GR", "HU", "IE", # EU
                  "IT", "LV", "LT", "LU", "MT", "NL", "PL", "PT", "RO", "SK", "SI", "ES", "SE"] # EU

article_count = {}
eu_order_rls = {}

for order in set(insect_articles["order"]):
    if order != "Isoptera": # actually a subclass...
        df = insect_articles[insect_articles["order"]==order]
        
        n = 0
        # count number of articles on this order that are in the EU
        for article in df.iloc[100:,].itertuples():
            if len(article.authorships) != 0:
                if len(article.authorships[0]["institutions"]) != 0:
                    if len(article.authorships[0]["institutions"][0]) != 0:
                        country = article.authorships[0]["institutions"][0]["country_code"]

            if country:
                if country in eu_codes:
                    n += 1
            country = None
            
        # calculate RLS
        article_count[order] = n
        eu_order_rls[order] = (n/species_per_order[order])*100

eu_order_rls
```

```python
article_count
```

These results are often worse than the RLIT found because often the first author is not European, but another author is. Another option is that the first institution of the author in question is not European. Following analysis...

```python
# filter for European (EU and pan-European) articles based on all institutes
eu_codes = ["AT", "BE", "BG", "HR", "CY", "CZ", "DK", "EE", "FI", "FR", "DE", "GR", "HU", "IE", # EU
                  "IT", "LV", "LT", "LU", "MT", "NL", "PL", "PT", "RO", "SK", "SI", "ES", "SE"] # EU
paneu_codes = ["IS", "LI", "NO", "CH", "AL", "ME", "MK", "RS", "TR", "AD", "BY", "BA", "MD", "MC", # pan-Europe
               "RU", "SM", "UA", "GB", "VA", "GE", "AM", "AZ"] # pan-Europe
eu_count = {}
eu_order_rls = {}
paneu_count = {}
paneu_order_rls = {}

for order in set(insect_articles["order"]):
    if order != "Isoptera": # actually a subclass...
        df = insect_articles[insect_articles["order"]==order]
        
        n_eu = 0
        n_pan = 0
        # count number of articles on this order that are in the EU
        is_eu = False
        is_pan = False
        
        for article in df.itertuples():
            # check every author
            for author in article.authorships:
                # check every affiliated institute
                for institute in author["institutions"]:
                    if institute:
                        country = institute["country_code"]
                        # EU?
                        if country in eu_codes:
                            is_eu = True
                            break # one European institute is enough
                        # pan-European?
                        elif country in paneu_codes:
                            is_pan = True
                            break
                            
                if is_eu or is_pan:
                    break # one European author is enough
            
            # count
            if is_eu:
                n_eu += 1
                n_pan += 1
            if is_pan:
                n_pan += 1
            
            is_eu = False
            is_pan = False

        # calculate Red List Score
        eu_count[order] = n_eu
        eu_order_rls[order] = (n_eu/species_per_order[order])*100
        paneu_count[order] = n_pan
        paneu_order_rls[order] = (n_pan/species_per_order[order])*100
```

```python
# filter for European (EU and pan-European) articles based on all institutes
eu_codes = ["AT", "BE", "BG", "HR", "CY", "CZ", "DK", "EE", "FI", "FR", "DE", "GR", "HU", "IE", # EU
                  "IT", "LV", "LT", "LU", "MT", "NL", "PL", "PT", "RO", "SK", "SI", "ES", "SE"] # EU
paneu_codes = ["IS", "LI", "NO", "CH", "AL", "ME", "MK", "RS", "TR", "AD", "BY", "BA", "MD", "MC", # pan-Europe
               "RU", "SM", "UA", "GB", "VA", "GE", "AM", "AZ"] # pan-Europe

eu_articles = []
paneu_articles = []

for article in insect_articles.itertuples():
    if article.order == "Isoptera": # actually a subclass...  
        break
    
    # check every author
    for author in article.authorships:
        stop = False
        # check every affiliated institute
        for institute in author["institutions"]:
            if institute:
                country = institute["country_code"]
                # european?
                if country in eu_codes:
                    eu_articles.append(list(article))
                    paneu_articles.append(list(article))
                    stop=True
                    break
                
                elif country in paneu_codes:
                    paneu_articles.append(list(article))
                    stop=True
                    break
        if stop:
            break

eu_articles = pd.DataFrame(eu_articles)
eu_articles = eu_articles.iloc[:,1:]
eu_articles.columns = insect_articles.columns

paneu_articles = pd.DataFrame(paneu_articles)
paneu_articles = paneu_articles.iloc[:,1:]
paneu_articles.columns = insect_articles.columns
```
