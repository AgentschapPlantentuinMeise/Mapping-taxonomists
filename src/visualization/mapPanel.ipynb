```python
import matplotlib.pyplot as plt
from PIL import Image
from pathlib import Path

# Define paths
root_dir = Path("D:/Documents/notebooks/TETTRIs-mapping-taxonomists9")  # Adjust if needed
fig_dir = root_dir / "reports" / "figures"

# Image paths
img1_path = fig_dir / "map_authors_europe.jpg"
img2_path = fig_dir / "map_authors_europe_relative.jpg"

# Load images
img1 = Image.open(img1_path)
img2 = Image.open(img2_path)

# Create figure with side-by-side subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

# Show images
axes[0].imshow(img1)
axes[0].axis('off')
axes[0].set_xlabel("Absolute number of authors", fontsize=20)

axes[1].imshow(img2)
axes[1].axis('off')
axes[1].set_xlabel("Relative to population (%)", fontsize=20)

plt.tight_layout()

output_path = fig_dir / "panel_map_authors_europe.jpg"
plt.savefig(output_path, dpi=300)
plt.show()

print(f"Combined panel saved to: {output_path}")

```
